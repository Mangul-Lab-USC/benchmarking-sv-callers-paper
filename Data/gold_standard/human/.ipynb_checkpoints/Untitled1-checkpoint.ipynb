```python
import pandas as pd

data_PCR=pd.read_csv('human_PCR.csv')

del_pcr_set=set()

for index, row in data_PCR.iterrows():
    x=int(row['SVStart'])
    y=int(row['SVEnd'])
    del_pcr_set.add((x,y))
    
print del_pcr_set
```

```python
import pandas as np
file=open("human_deletions_nanopore_PCR_combined.csv","w")
file.write("SVChrom,SVStart,SVEnd,SVSpan,type\n")

for index, row in data_nanopore.iterrows():
    chr=row['SVChrom']
    x=int(row['SVStart'])
    y=int(row['SVEnd'])
    l=int(row['SVSpan'])
    if (x,y) in del_pcr_set:
        file.write(chr+","+str(x)+","+str(y)+","+str(l)+",PCR-validated")
        file.write("\n")
    else:
        file.write(chr+","+str(x)+","+str(y)+","+str(l)+",Nanopore-validated")
        file.write("\n")

file.close()        
        
        
data_nanopore2=np.read_csv('human_deletions_nanopore_PCR_combined.csv')


data_nanopore2.head()
```
