```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cpu_mouse.csv") 
```

```python
data
```

```python
# Varscan takes the longest, the two samples show similar behavior
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="sample_name", y="time",data=data,hue='Tool',aspect=1.8,kind='bar')

g.set(xlabel='Tools', ylabel='Run time (seconds)')
g=sns.despine()
plt.show(g)
```

```python
# Varscan takes the most memory, the two samples show similar behavior
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="sample_name", y="memory",data=data,hue='Tool',aspect=1.8,kind='bar')

g.set(xlabel='Tools', ylabel='Memory (MB)')
g=sns.despine()
plt.show(g)
```

```python
#Time taken by 1x is negligible as compared to full coverage sample 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="time",data=data,hue='coverage',aspect=1.8,kind='bar')

g.set(xlabel='Tools', ylabel='Run time (seconds)')
g=sns.despine()
plt.show(g)
```

```python
#Memory for full coverage and 1x is comparable 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="memory",data=data,hue='coverage',aspect=1.8,kind='bar')

g.set(xlabel='Tools', ylabel='Memory (MB)')
g=sns.despine()
plt.show(g)
```
